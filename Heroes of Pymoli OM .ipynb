```python
 # Dependencies
import pandas as pd 
import numpy as np 
```

```python
# Save file path to variable
file_to_load = "purchase_data.csv"
```

```python
# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
#Display total number of players 
unique_players= len(purchase_data["SN"].unique())
total_players = pd.DataFrame({"Total Players": [unique_players]})
total_players
```

```python
#Run basic calculations to obtain number of unique items, average price, etc.
unique_items = len(purchase_data["Item ID"].unique())

avg_price = purchase_data["Price"].mean()

num_purchases = purchase_data["Purchase ID"].count()

tot_revenue = purchase_data["Price"].sum()
```

```python
#Create a summary data frame to hold the results
summary = pd.DataFrame({"Number of Unique items": [unique_items], "Average Price": [avg_price], 
               "Numer of Purchases": [num_purchases], "Total Revenue": [tot_revenue]})
summary["Average Price"] = summary["Average Price"].map("${:.2f}".format)
summary["Total Revenue"] = summary["Total Revenue"].map("${:.2f}".format)
```

```python
#Display the summary data frame (Optional: give the displayed data cleaner formatting)
summary
```

```python
#Remove duplicate SN in OG dataframe 
no_duplicates = purchase_data.sort_values("SN")
no_duplicates.drop_duplicates(subset ="SN", keep = "first", inplace = True)
```

```python
#Percentage and Count of Male Players
male_players = no_duplicates.loc[purchase_data["Gender"]== "Male"]
male_count = len(male_players)
male_per = (male_count / unique_players)*100
```

```python
#Percentage and Count of Female Players
female_players = no_duplicates.loc[purchase_data["Gender"]== "Female"]
female_count = len(female_players)
female_per = (female_count / unique_players)*100
```

```python
#Percentage and Count of Other / Non-Disclosed
other_players = no_duplicates.loc[purchase_data["Gender"]== "Other / Non-Disclosed"]
other_count = len(other_players)
other_per = (other_count / unique_players)*100
```

```python
#Create and display the summary data frame (Optional: give the displayed data cleaner formatting)
#gender_demographics = pd.DataFrame({"Total Count": [male_count, female_count, other_count], 
                                    #"Percentage of Players": [male_per, female_per, other_per]}
gender_demographics = pd.DataFrame({"": ["Male", "Female", "Other" ], 
                                    "Total Count": [male_count, female_count, other_count], 
                                    "Percentage of Players": [male_per,female_per,other_per]})

gender_demographics["Percentage of Players"] = gender_demographics["Percentage of Players"].map("{:.2f}".format)

gender_demographics.set_index("")

gender_demographics
```

```python
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

total_females = purchase_data.loc[purchase_data["Gender"]== "Female"]
total_females_count = total_females["Purchase ID"].count()
avg_female_purchase_price = total_females["Price"].mean()
female_sum = total_females["Price"].sum()
female_total = total_females["Price"].sum()
female_total_purchase = (female_total / female_count)

total_males = purchase_data.loc[purchase_data["Gender"]== "Male"]
total_males_count = total_males["Purchase ID"].count()
avg_male_purchase_price = total_males["Price"].mean()
male_sum = total_males["Price"].sum()
male_total = total_males["Price"].sum()
male_total_purchase = (male_total / male_count)

total_other = purchase_data.loc[purchase_data["Gender"]== "Other / Non-Disclosed"]
total_other_count = total_other["Purchase ID"].count()
avg_other_purchase_price = total_other["Price"].mean()
other_sum = total_other["Price"].sum()
other_total = total_other["Price"].sum()
other_total_purchase = (other_total / other_count)

purchase_gender = pd.DataFrame({"": ["Female", "Male", "Other"], "Purchase Count": [total_females_count, total_males_count, total_other_count], 
                                "Average Purchase Price": [avg_female_purchase_price, avg_male_purchase_price, avg_other_purchase_price],
                               "Total Purchase Value": [female_sum, male_sum, other_sum], 
                               "Avg Total Purchase by Person": [female_total_purchase, male_total_purchase, other_total_purchase]})

purchase_gender["Average Purchase Price"] = purchase_gender["Average Purchase Price"].map("${:.2f}".format)
purchase_gender["Total Purchase Value"] = purchase_gender["Total Purchase Value"].map("${:.2f}".format)
purchase_gender["Avg Total Purchase by Person"] = purchase_gender["Avg Total Purchase by Person"].map("${:.2f}".format)

purchase_gender.set_index("")
purchase_gender
```

```python
no_duplicates_group = no_duplicates.groupby("Age Group")

group_count = no_duplicates_group["SN"].count()

group_percentage = (group_count / unique_players)*100

age_demographics = pd.DataFrame({"Total Count": group_count, "Percentage of Players": group_percentage})

age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].map("{:.2f}".format)

age_demographics
```

```python
#Establish bins for ages
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_names)

#purchase_data.head()
```

```python
purchase_data_group = purchase_data.groupby("Age Group")
age_group_count = purchase_data_group["SN"].count()
age_group = len(age_group_count)
avg_purchase_price_group = purchase_data_group["Price"].mean()
total_purchase_value_group = purchase_data_group["Price"].sum()
group_total_purchase = ( total_purchase_value_group/ group_count)

purchasing_analysis = pd.DataFrame({"Purchase Count": age_group_count, "Average Purchase Price": avg_purchase_price_group, 
                                    "Total Purchase Value": total_purchase_value_group, "Avg Total Per Person" : group_total_purchase})

purchasing_analysis["Average Purchase Price"] = purchasing_analysis["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis["Total Purchase Value"] = purchasing_analysis["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis["Avg Total Per Person"] = purchasing_analysis["Avg Total Per Person"].map("${:.2f}".format)

purchasing_analysis
```

```python
#Total Spenders
total_purchase = purchase_data.groupby("SN")
total_spenders_count = pd.DataFrame(total_purchase["Purchase ID"].count())

total_price = pd.DataFrame(total_purchase["Price"].mean())
total_price_1 = total_price.sort_values(by="Price", ascending=False)

sum_purchase = pd.DataFrame(total_purchase["Price"].sum())
sum_purchase_1 = sum_purchase.sort_values(by="Price", ascending=False)


total_spenders_count["Average Purchase Price"] = total_price_1
total_spenders_count["Total Purchase Value"] = sum_purchase_1

total_spenders_count = total_spenders_count.sort_values(by="Total Purchase Value", ascending=False)

total_spenders_count["Average Purchase Price"] = total_spenders_count["Average Purchase Price"].map("${:.2f}".format)
total_spenders_count["Total Purchase Value"] = total_spenders_count["Total Purchase Value"].map("${:.2f}".format)


total_spenders_count.head(5)


#Total_Spenders
```
